```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("/content/sample_data/fuel_econ.csv")

plt.figure(figsize=(10, 6))
sns.histplot(df['comb'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Combined Fuel Economy (MPG)')
plt.xlabel('Combined Fuel Economy (MPG)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Histogram Interpretation:**

The histogram shows the distribution of combined fuel economy (MPG) across vehicles in the dataset.

Most vehicles fall within the **20 to 30 MPG range**, indicating standard efficiency levels.  

There are fewer vehicles at the low end (<15 MPG), likely representing heavy-duty or performance vehicles, and fewer at the high end (>35 MPG), likely representing hybrid or electric vehicles.  

This distribution suggests that the dataset mostly consists of moderately fuel-efficient vehicles.

**Heatmap Interpretation:**

This heatmap shows the correlation between various numerical features in the vehicle dataset.

- **Strong Negative Correlation (dark blue):**
  - `co2` vs `comb`, `city`, `highway`, `UCity`, `UHighway`: These are strongly negatively correlated. As fuel efficiency (MPG) increases, CO2 emissions decrease — which is expected.
  - `ghgScore` and `feScore` vs `co2`: Also strongly negative. Vehicles with higher environmental scores emit less CO2.

- **Strong Positive Correlation (dark red):**
  - `city`, `highway`, `comb`, `UCity`, and `UHighway`: These fuel economy metrics are highly correlated with each other, indicating consistent performance measures.
  - `cylinders` and `displ`: More cylinders usually mean larger engine displacement.

- **Low or No Correlation (white/light colors):**
  - `id`, `year`, and `pv2/pv4` show minimal correlation with most other features, suggesting they have less impact on performance or emissions.

### Overall:
The heatmap confirms expected relationships: fuel-efficient vehicles pollute less and get higher environmental scores. Engine size and cylinder count are linked to higher emissions and lower fuel economy.
