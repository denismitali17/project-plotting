```python
df = pd.read_csv("fuel_econ.csv")

plt.figure(figsize=(10, 6))
sns.histplot(df['comb'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Combined Fuel Economy (MPG)')
plt.xlabel('Combined Fuel Economy (MPG)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Histogram Interpretation:**

The histogram shows the distribution of combined fuel economy (MPG) across vehicles in the dataset.

Most vehicles fall within the **20 to 30 MPG range**, indicating standard efficiency levels.  

There are fewer vehicles at the low end (<15 MPG), likely representing heavy-duty or performance vehicles, and fewer at the high end (>35 MPG), likely representing hybrid or electric vehicles.  

This distribution suggests that the dataset mostly consists of moderately fuel-efficient vehicles.


```python
df = pd.read_csv("fuel_econ.csv")

plt.figure(figsize=(8, 5))
plt.hist(df['co2'], bins=30, color='purple', edgecolor='white', linewidth=0.7)
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Count of Vehicles')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The histogram of CO₂ emissions shows how CO₂ output (in grams per mile) is distributed across vehicles in the dataset.

- The distribution appears **right-skewed**, indicating that most vehicles emit moderate levels of CO₂, while fewer emit very high levels.
- The **peak** of the histogram is centered around a **moderate emission range** (likely 250–350 g/mile), suggesting that many vehicles in the dataset fall within this efficiency bracket.
- The **longer tail** to the right indicates that there are some vehicles with **very poor fuel economy**, resulting in higher CO₂ output.
- This skewness can help in identifying outliers or extreme polluters and inform fuel economy policies.


```python
plt.figure(figsize=(6, 5))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, cmap='RdBu_r', annot=False, linewidths=0.5, center =0)
plt.title('Correlation Heatmap of Vehicle Features')
plt.xticks(rotation=45)
plt.show()
```
